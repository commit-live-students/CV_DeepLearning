## 1.2 Digital Image representation:
Computer vision being the business of developing algorithms that enable computers to intepret images the way humans do, it would have been more convinient if we had known our brain stores and process images.  Let's start by looking at how images are represented in computers. We freeze a moment of time and capture the 3D world around us and project them in to a 2D screen within the camera. Images are represented in a 2D grid of fixed size and each point in the grid stores three numerical values correspoding to Red,Green and Blue components of the scene.<br>

<img src = "../images/computer_cat.png">

In this example, the cat image is 248 pixels wide, 400 pixels tall, and has three color channels Red,Green,Blue (or RGB for short). Therefore, the image consists of 248 x 400 x 3 numbers, or a total of 297,600 numbers. Each number is an integer that ranges from 0 (black) to 255 (white). That's already a quarter million numbers for a relatively smaller images and if we use this in this form our learning algorithms might take forever to learn due to computational complexity. Hence there is a need to transform this array of numbers in to a form that keeps our learning algorithms tractable and also aid them in their task.

## 1.3 Desirable Properites
Objects in real world could exhibit variations and the image representation we chose should be roboust to such variations. We list below some of the variations,<br>

* ViewPoint variation - Different  perspectives of the camera might result in different images.
* Scale  - The same object could exist in different sizes .
* Deformation - Non-rigid objects appears to exist in many different possible forms.
* Occulsion - Sometimes only a part of the object appears.
* Illumination condition - Depending on the lighting condition, the captured image varies.

<img src ="../images/challenges.jpeg" >

Due to such variations the pixels we observe might have different values inspite of having the same real world object in it. To achieve our goal of  visual perception (Object recognition, Object localisation etc..) the image representation should have certain desirable properities.<br>

* Ignore irrelevant variations - The representation should not vary with environmental changes such as lighting or any other change that doesn't affect the object identity. For example when our task is to distinguish a bat from a ball, the representation needn't change when the color of the object changes.
* Capture discriminatory information - The representation should capture all distinguishing variations - for example when our task is to distinguish a red ball from yellow ball, the representation should vary for each of these colors.


Thus a good representation exhibits invariance to external changes and discrimination between object and other components of the image. It is important to note that the representation can be judged only based on the task it is employed for.<br>

## 2.1 Local and Global features
Over the years computer vision researchers have come up with numerous image transformations that can be used as representations. The transformed image is generally called as features that describe the content of the image - it is an alternative representation of the image. These features are broadly classified in to local or global features based on how they operate on a image.<br>
Local features operate on a small subset of the image and describe only that region (Example:Edge detectors),whereas global features describe the entire image (example: Histograms).

## 2.3 Histogram of oriented gradients

Histogram of Oriented Gradients(HOG) is a transformation which characterises a given image by counting the number of times different gradient occurs in a image. It's a feature that is suited to describe shapes and was first proposed for the task of human detection.

Divide the image into small connected regions called cells, and for each cell compute a histogram of gradient directions or edge orientations for the pixels within the cell.
Discretize each cell into angular bins according to the gradient orientation.
Each cell's pixel contributes weighted gradient to its corresponding angular bin.
Groups of adjacent cells are considered as spatial regions called blocks. The grouping of cells into a block is the basis for grouping and normalization of histograms.
Normalized group of histograms represents the block histogram. The set of these block histograms represents the descriptor.


Below are the steps to compute HOG,

1. An optional first step where intensity normalisation is performed either by taking log(x) of each pixel or square root of each pixel.

2. The next step  is to divide the image into blocks (for example 1616 pixels). Each block is divided by small regions, called cells (for example 44 pixels). The cells can overlap.

3. Compute gradient of the image in x and y direction and compute the gradient magnitude by summing up the gradient in x and y direction. Also compute the orientation of gradient for every pixel.

4. Construct a histogram for every cell,that counts the number of pixels in a given gradient orientation interval.For example if we set number of bins in the histogram to 6 , then each bin represents 30 degrees (180/6).

5. Now group the cells in a block and normalise the histogram. This step ensures that the transformation is invariant to intensity. We have now obtained a normalised histogram for every block.

6. The set of all block histograms defines the HOG descriptor for a given image. 

HOG computation has following configurable parameters,

* Size of the cell- pixels per cell
* Size of the block - cells per block
* Size of the histogram - number of bins

Let us use scikit-image to compute histogram of an image.

```python
from skimage import exposure
from skimage import feature


im = data.camera()
(H, hogImage) = feature.hog(im, orientations=9, pixels_per_cell=(8, 8),
                cells_per_block=(2, 2), transform_sqrt=True,block_norm="L1",
                visualise=True)
hogImage = exposure.rescale_intensity(hogImage, out_range=(0, 255))
hogImage = hogImage.astype("uint8")

plt.imshow(hogImage)
```

## 2.4 Image classsification using HOG

Lets extract HOG features and use them to train an image classifier to identify digits.

```python


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



SZ = 20
CLASS_N = 10

def split2d(img, cell_size, flatten=True):
    h, w = img.shape[:2]
    sx, sy = cell_size
    cells = [np.hsplit(row, w//sx) for row in np.vsplit(img, h//sy)]
    cells = np.array(cells)
    if flatten:
        cells = cells.reshape(-1, sy, sx)
    return cells

def load_digits(fn):
    digits_img = cv2.imread(fn, 0)
    digits = split2d(digits_img, (SZ, SZ))
    labels = np.repeat(np.arange(CLASS_N), len(digits)/CLASS_N)
    return digits, labels

def deskew(img):
    m = cv2.moments(img)
    if abs(m['mu02']) < 1e-2:
        return img.copy()
    skew = m['mu11']/m['mu02']
    M = np.float32([[1, skew, -0.5*SZ*skew], [0, 1, 0]])
    img = cv2.warpAffine(img, M, (SZ, SZ), flags=cv2.WARP_INVERSE_MAP | cv2.INTER_LINEAR)
    return img




def preprocess_simple(digits):
    return np.float32(digits).reshape(-1, SZ*SZ) / 255.0
```

```python
print('Loading digits from digits.png ... ')
# Load data.
digits, labels = load_digits('../images/digits.png')
# print(digits[0])


plt.figure(figsize=(5,5))

import random 


f = plt.figure(figsize=(10,7))

for i in range(20):

  
  index = random.randint(0,digits.shape[0])
  plt.subplot( 5, 4, i+1 )
  plt.axis('off')
  plt.title(labels[index])

  plt.imshow(digits[index], cmap="gray")
  
  
plt.tight_layout( w_pad=0.5, h_pad=1.0)

plt.show()
```

```python
def compute_hog(image,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2)):
    hog= feature.hog(im, orientations=orientations, pixels_per_cell=pixels_per_cell,
                cells_per_block=cells_per_block, transform_sqrt=True,block_norm="L1")
    return hog
    
```

```python


print('Calculating HoG descriptor for every image ... ')
hog_descriptors = []
for i,img in enumerate(digits):
    hog_descriptors.append(compute_hog(img))
hog_descriptors = np.squeeze(hog_descriptors)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(hog_descriptors, labels, test_size = 0.3, random_state = 10)
print("Training size", X_train.shape[0])
print("Testing Size" , X_test.shape[0])
```

```python
from sklearn.svm import SVC

model = SVC(kernel = "rbf", C = 50, gamma =0.50625)

model.fit(X_train, y_train)
preds = model.predict(X_test)


from sklearn.metrics import confusion_matrix

mat = confusion_matrix(y_test, preds, labels = np.arange(0,10))
print("Confusion matrix")
print(mat)

print("Accuracy:" , model.score(X_test, y_test))
```

We get  98% accuracy on a multi-class classification with HOG as feature. This illustrates that HOG is good at discerning shapes. But in most real-world applications there are other dimensions like apperance,texture that needs to be captured as well. Its very difficult to engineer a feature that can capture all the properties of an image that aids in the chosen task. It might be useful to learn the features rather than engineering it and that's what neural networks do.
