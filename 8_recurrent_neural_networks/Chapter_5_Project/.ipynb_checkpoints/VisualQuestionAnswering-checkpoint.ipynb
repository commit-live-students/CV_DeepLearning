## Visual Question Answering (VQA)

VQA is one of the appliocation from natural language processing domain. Here you will present an image to the software along with a question related to that image and software generate an output on the basis of image content and question.

This process uses leverage of CNN for image feature extraction to understand what is happening in the image. It uses RNN to extract information from question. Finally we combine both the information using an aggregator network as similar to our image captionaing framework. Aggregator network generate an Answer to the asked question. ### Textual feature extraction

We will use Long Short Term Memory (LSTM) units to create this network. And you already aware that it is a kind of recurrent neural network which have been used in sequence based operations. As these are special kind of networks which have small memory units along with weights. It helps to learn the previous time steps of the input. That’s why these are best suited for natural language processing tasks such as text generation, machine translation, speech recognition etc. 

Now before creating a LSTM network, we need to do lot of pre-processing on input text, we need to create a concise vocabulary so that our network don’t get a hard time during training, we need to remove capital letters, punctuations etc. from the text data, it will help us to design a simple yet powerful LSTM model.

After getting both kinds of information processors, now we need to get ahead with our aggregator model.

### Aggregator model

This network will be a fully connected neural network which will process combined information obtained from both the feature extractors. Input of the aggregator network is a vector which consists of visual information produced by VGG network and textual information processed by LSTM network, this network will generate softmax probabilities, which eventually will give us a word vector in the end and it will help to generate caption for the image.

## Text processing

We will start with creating separate dictionaries for questions and answers so that we can clean these descriptions. We will remove all the punctualtions from the descriptions and convert all the letters into lower case. It will help us to create a concise vocabulary to be learn by neural network. 

First we will write a function to load the text document.

```python
#String will handle text processing operations
import string

data_dir = 'D:/GoogleDrive/TechDocuments/Webinars/GrayAtom/Chapter_5_Project/VQA_Data/'

def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

Following function will create 3 dictionaries for Questions, Answers and combination of both of them. We will clean this dictionaries and use them later in the framework.

```python
def text_dictionary(doc):
    
    # map questions for each image
    question_mapping = dict()
    answer_mapping = dict()
    mapping = dict()
    lines = doc.split('\n')    
    for n in range(len(lines)-1):
    
        # load an image from file
        line = lines[n].split(';')
        image_id = line[0]
        question = line[1]
        answer = line[2]
        
        # Remove file extension from image name 
        image_id = image_id.split('.')[0]
        
        # Create key if not present
        if image_id not in question_mapping:
            question_mapping[image_id] = list()
            answer_mapping[image_id] = list()
            mapping[image_id] = list()
        
        # Insert descriptions as key value in respected keys
        question_mapping[image_id].append(question)
        answer_mapping[image_id].append(answer)
        mapping[image_id].append(question+' '+answer)
        
    return mapping,question_mapping,answer_mapping
```

Now we will write a function to clean the descriptions loaded in the dictionaries. Following function will do it for us.

```python
#Following function will clean the pre process the descriptions
def clean_descriptions(descriptions):
    
    #DESCRIPTIONS: is the dictionary contains image descriptions
    
    # We prepare a translation table which contains all the punctuation
    table = str.maketrans('', '', string.punctuation)
    
    # Let's iterate with different descriptions
    for key, desc_list in descriptions.items():
        
        # Here we will work on each description
        for i in range(len(desc_list)):
            
            # Extract description
            desc = desc_list[i]
            
            # Split it into words (tokenize)
            desc = desc.split()
            
            # Convert to all letters in lower case
            desc = [word.lower() for word in desc]
            
            # Remove punctuation from each token
            desc = [w.translate(table) for w in desc]                                    
            
            # Join all the words again to store them
            desc_list[i] =  ' '.join(desc)
```

After cleaning the descriptions we need to have one more function to save them in the text formet. We will do it with the help of following function.

```python
# save descriptions to file, one per line
def save_descriptions(descriptions, filename):
    
    #DESCRIPTIONS: Cleaned descriptions dictionary
    #FILENAME: filename to store description onto disk
    
    #Create a list to store different lines
    lines = list()
    
    #Iterate to each descriptions
    for key, desc_list in descriptions.items():
        
        #Put the description with image name in different line
        for desc in desc_list:
            lines.append(key + ' ' + desc)
    
    #Join all the lines in single document 
    data = '\n'.join(lines)
    
    #Open a file with defined filename with write permissions  
    file = open(filename, 'w')
    
    #Write the data into the file
    file.write(data)
    
    #Close the file
    file.close()
```

So we have all the helping functions for pre-processing of our text data now let's run these functions one by one and save the cleaned descriptions to the disk using following lines

```python
# Load selected annotations
annotation_file = 'annotations.txt'
txt_file = data_dir+annotation_file
doc = load_doc(txt_file)
mappings,qmappings,amappings = text_dictionary(doc)

# Clean descriptions
clean_descriptions(mappings)
clean_descriptions(qmappings)
clean_descriptions(amappings)

#Save descriptions to file
save_descriptions(mappings, data_dir+'descriptions.txt')
save_descriptions(qmappings, data_dir+'q_descriptions.txt')
save_descriptions(amappings, data_dir+'a_descriptions.txt')
```

Once text pre-processing done; this is the time to create vocabulary to be learn by our network. We will create a Tokenizer from cleaned description. Tokenizer will convert charecter data into numerical sequences. This numerical sequences will be used as text features to be feed into LSTM network. 

To create a Tokenizer we will:

- Load cleaned descriptions (File contains Questions and Answers both).
- Extract Text line by line.
- Fit a Tokenizer on extracted lines to build the vocabulary.

We will also add end of sequeces flag to each answer. It will help us to know where to stop the text generation. We will write helping function for all these three steps in following cell.

```python
#Pickle will be use to load and store the data
from pickle import dump

#Keras will help us in text related operations too
from keras.preprocessing.text import Tokenizer

# Load clean descriptions into memory
def load_clean_descriptions(doc, dataset=None):
    
    #FILENAME: file name of cleaned descriptions
    #DATASET: is file with image names created by load_set    
    
    #Create empty list to store sub set of description 
    descriptions = dict()
    
    #Let's iterate with different lines
    for line in doc.split('\n'):
        
        # Create words by splitting lines
        tokens = line.split()
        
        # Split id from description
        image_id, image_desc = tokens[0], tokens[1:]
                    
        # Put start and end flag to each description
        desc = ' '.join(image_desc)
        
        # Create list
        if image_id not in descriptions:
            descriptions[image_id] = list()    
        
        # Store modified description
        descriptions[image_id].append(desc)
    
    #Return descriptions 
    return descriptions

# Convert a dictionary of clean descriptions to a list of descriptions
def to_lines(descriptions):
    
    #DESCRIPTIONS: descriptions dictionary
    
    #Create an empty list to store descriptions
    all_desc = list()
    
    #Iterate through the disctionary keys and extract all descriptions
    for key in descriptions.keys():
        
        #Append all description one by one
        [all_desc.append(d) for d in descriptions[key]]
    
    #Return the description list
    return all_desc

# Fit a tokenizer given caption descriptions
def create_tokenizer(descriptions):
    
    #DESCRIPTIONS: descriptions dictionary
    
    #Convert all descriptions in the list form
    lines = to_lines(descriptions)
    
    # Add an end flag to vocab
    lines += ['endseq']
    
    #Create tokenizer object from Keras
    tokenizer = Tokenizer()
    
    #Here we will fit a tokenizer on text data     
    tokenizer.fit_on_texts(lines)
    
    #Return tokenizer
    return tokenizer
```

Let's use these function to create a Tokenizer.

```python
# Load the description file
txt_file = 'descriptions.txt'
txt_file = data_dir+txt_file

descriptions = load_doc(txt_file)

# Convert description file into dictionary formet
descriptions = load_clean_descriptions(descriptions)

#Prepare tokenizer
#We will call create tokenizer here for training descriptions
tokenizer = create_tokenizer(descriptions)

#This is our vocabulary size
vocab_size = len(tokenizer.word_index) + 1
print('Vocabulary Size: %d' % vocab_size)

#Store the tokenizer for re use 
dump(tokenizer, open(data_dir+'tokenizer.pkl', 'wb'))
```

Great We have a vocabulary of 1935 words. Now our next task will be creating a sequences to feed to the network. We will create sequeneces such that input sequece will be Question and output sequneces will be answer for the question. Now if the answer contains more than one word we will repeat the question with addition of previous answer. Following is an example of this process.

    Q: What colors are the umbrellas
    A: yellow and blue
    
Our sequeces will be:

    Image                Input                                     Output                         
    im           What colors are the umbrellas                     yellow
    im           What colors are the umbrellas yellow              and
    im           What colors are the umbrellas yellow and          blue
    im           What colors are the umbrellas yellow and blue     endseq
    
We will Do this in following steps:

- Load Image features and store in a list.
- Load cleaned questions and answers.
- Create input and output sequences from loaded questions and answers and store it into different lists.

We will do above three steps in following cell

```python
#OS will help us to get files from the directory
from _pickle import load
from os import listdir

from keras.preprocessing.sequence import pad_sequences
from keras.utils.np_utils import to_categorical

import numpy as np

# load photo features
def load_image_features(filename):
    
    #FILENAME: filename containing image features    
    #load all features
    
    all_features = load(open(filename, 'rb'))    
    return all_features

# Load clean descriptions into memory
def load_clean_descriptions(filename):
    
    #FILENAME: file name of cleaned descriptions
    #DATASET: is file with image names created by load_set
    
    # Load cleaned descriptions
    doc = load_doc(filename)
    
    #Create empty list to store sub set of description 
    descriptions = dict()
    
    #Let's iterate with different lines
    for line in doc.split('\n'):
        
        # Create words by splitting lines
        tokens = line.split()
        
        # Split id from description
        image_id, image_desc = tokens[0], tokens[1:]            
                    
        # Create list
        if image_id not in descriptions:
            descriptions[image_id] = list()
        
        # Put start and end flag to each description
        desc = ' '.join(image_desc)
        
        # Store modified description
        descriptions[image_id].append(desc)
    
    #Return descriptions 
    return descriptions

def create_data(q_descriptions,a_descriptions,image_features,vocab_size):
    
    #Let's start with creating empty list to store our sequences
    #for both input features and output word for each input
    X1, X2, y = list(), list(), list()
    max_length = 25
    
    for key in q_descriptions.keys():
    
        questions = q_descriptions[key]
        answers = a_descriptions[key]
        images = image_features[key]
        
        for ques,ans in zip(questions,answers):
            ans +=' '+'endseq'
            q_seq = tokenizer.texts_to_sequences([ques])[0]
            a_seq = tokenizer.texts_to_sequences([ans])[0]            
                  
            for a_word in a_seq:
                                
                in_seq = pad_sequences([q_seq],maxlen=max_length)[0]
                out_seq = to_categorical([a_word], num_classes=vocab_size)[0]
        
                X1.append(images[0])
                X2.append(in_seq)
                y.append(out_seq)
                
                q_seq += [a_word]
                
    #Return sequences            
    return np.array(X1), np.array(X2), np.array(y)
```

Great! Now we will create sequences in following lines.

```python
# Load cleaned questions
q_descriptions = data_dir+'q_descriptions.txt'

# Load Cleaned answers
a_descriptions = data_dir+'a_descriptions.txt'

# Load tokenizer to convert words into integer form
token  = data_dir+'tokenizer.pkl'
image_features = data_dir+'features.pkl'
tokenizer = load(open(token, 'rb'))

#This is our vocabulary size
vocab_size = len(tokenizer.word_index) + 1
print('Vocabulary Size: %d' % vocab_size)

# Load Image features and descriptions
q_descriptions = load_clean_descriptions(q_descriptions)
a_descriptions = load_clean_descriptions(a_descriptions)
image_features = load_image_features(image_features)

# Create sequences
im,ques,ans = create_data(q_descriptions,a_descriptions,image_features,vocab_size)

# Store sequences onto the disk for further usage
np.save(data_dir+'image_features.npy',im)
np.save(data_dir+'ques_features.npy',ques)
np.save(data_dir+'ans_features.npy',ans)

print("Done")
```

All looks great! We have have Inputs and Outputs ready to feed into the neural network. Now this is the time to create our aggregator network and train it with the sequences.

We can summarize our process as;

-	Image Feature Extractor. This is a 16-layer VGG model pre-trained on the ImageNet dataset. We have pre-processed the photos with the VGG model (without the output layer) and will use the extracted features predicted by this model as input.
-	Sequence Processor. This is a word embedding layer for handling the text input, followed by a Long Short-Term Memory (LSTM) recurrent neural network layer.
-	Decoder (for lack of a better name). Both the feature extractor and sequence processor output a fixed-length vector. These are merged together and processed by a dense layer to make a final prediction.

The Image Feature Extractor model expects input photo features to be a vector of 4,096 elements. These are processed by a dense layer to produce a 256 element representation of the photo.

The Sequence Processor model expects input sequences with a pre-defined length (34 words) which are fed into an Embedding layer that uses a mask to ignore padded values. This is followed by an LSTM layer with 256 memory units.

Both the input models produce a 256 element vector. Further, both input models use regularization in the form of 25% dropout. This is to reduce overfitting the training dataset, as this model configuration learns very fast.

The Decoder model merges the vectors from both input models using an addition operation. This is then fed to a Dense 256 neuron layer and then to a final output Dense layer that makes a softmax prediction over the entire output vocabulary for the next word in the sequence.

Let's create our model and train it on the sequences.

```python
#Follwing will help us to create and loading Models 
from keras.models import Model, load_model
from keras.layers import Input, Dropout,Dense,Embedding,LSTM,add

#Callbacks will be use to choose model configurations
from keras.callbacks import ModelCheckpoint

#Pickle will be use to load and store the data
from _pickle import load

# Define the captioning model
def define_model(vocab_size, max_length=25):
    
    #VOCAB_SIZE: Number of unique words
    #MAX_LENGTH: maximum length of sequence
    
    # Feature extractor model
    inputs1 = Input(shape=(4096,))
    fe1 = Dropout(0.25)(inputs1)
    
    #Compress the size of 4096 feature vector to 256
    fe2 = Dense(256, activation='relu')(fe1)
    
    # Sequence model
    inputs2 = Input(shape=(max_length,))
    
    # Add a word embedding layer to learn vector representation
    # of that word
    se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)
    se2 = Dropout(0.25)(se1)
    
    # Add a LSTM layer to process the word vector
    # LSTM will predict next word of the sequence
    # Here it is independent of image. 
    se3 = LSTM(256)(se2)
    
    # Decoder model
    # This model will responsible to predict a word against
    # an image and a sequence.
    
    # First we will combine a image and a sequence
    # predicted by LSTM
    decoder1 = add([fe2, se3])
    
    # And we will process the combined information here
    decoder2 = Dense(256, activation='relu')(decoder1)
    outputs = Dense(vocab_size, activation='softmax')(decoder2)
    
    # Here we will define input and output variable layers of our model
    model = Model(inputs=[inputs1, inputs2], outputs=outputs)
    
    #We will compile our model with cross entropy loss and
    #Adaptive momentum variant of stochastic gradient descent
    model.compile(loss='categorical_crossentropy', optimizer='adam')
    
    # Let's our model parameters
    print(model.summary())
    
    #Plot model to visualize it and store on to local disk
    #plot_model(model, to_file=data_dir+'model.png', show_shapes=True)
    
    #Return the created model
    return model
```

All looks well. Now let's train the defined model in following lines.

```python
#Load tokenizer to get vocabulary size
token  = data_dir+'tokenizer.pkl'
tokenizer = load(open(token, 'rb'))
vocab_size = len(tokenizer.word_index) + 1
print('Vocabulary Size: %d' % vocab_size)

# Load all the created input and outputs
image_features = np.load(data_dir+'image_features.npy')
ques_features = np.load(data_dir+'ques_features.npy')
ans_features = np.load(data_dir+'ans_features.npy')

# Define the VQA model
vqa_model = define_model(vocab_size)

# Store our model on the disk
filepath = data_dir+'vqa_model-ep{epoch:03d}-loss{loss:.3f}.h5'
checkpoint = ModelCheckpoint(filepath, monitor='loss', 
                             verbose=1, save_best_only=True, mode='min')

# Fit model our model for 20 epochs 
vqa_model.fit([image_features, ques_features],ans_features,batch_size=128,epochs=20, 
          verbose=2, callbacks=[checkpoint])

print("Done with training")
```

Once the training done; Its time to test our model. We will select an input image from our data set and try to get answer on that. Following function will generate an answer for the input question it will use a helping function for converting predicted integer into charecter formet. 

```python
# Map an integer to a word
def word_for_id(integer, tokenizer):
    
    #INTEGER: Maximum probability index
    #TOKENIZER: tokenizer for word mapping
    
    #Search the word using index value  
    for word, index in tokenizer.word_index.items():
        
        #If found a match return the word
        if index == integer:
            return word
    #If no match found retrun NONE    
    return None

# Generate a description for an image
def generate_desc(model, tokenizer, photo, question, max_length=25):        
    
    #MODEL: trained caption generator model
    #TOKENIZER: to map word to integer and integer to word
    #PHOTO: Image features
    #MAX_LENGTH: maximum sequence length with model trained earlier        
    
    # Our initial word for the sequence 
    in_text = question
    answer = ' '
    # Start iterating for the sequence    
    for i in range(max_length):
                
        # Integer encode input sequence
        sequence = tokenizer.texts_to_sequences([in_text+' '+answer])[0]
        
        # Pad input 
        sequence = pad_sequences([sequence], maxlen=max_length)
        
        # Predict next word with image and previous sequence as input
        yhat = model.predict([photo,sequence], verbose=0)
        
        # Get maximum probability index
        yhat = np.argmax(yhat)
        
        # Convert index into word using tokenizer
        word = word_for_id(yhat, tokenizer)
        
        # Stop is there is no predictions
        if word is None:
            break
        
        if word == 'endseq':
            break
        
        # Append predicted words to the sequence
        answer += ' ' + word
        
    #Return the generated caption
    return answer
```
