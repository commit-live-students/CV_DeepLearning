# 2.1 What is an Autoencoder (AE)

Let's start by breaking the name 'Autoencoder' into two part; We will get 'Auto' and 'Encoder'. Well thats what it is. It is an algorithm which can encode the given data automatically! (Of course still you need to write the code). Basically this is a classic machine learning algorithm which shares the domain with artificial neural networks. It first convert the data into the latent space and then from this latent space it tries reconstruct the data into its original form.

Converting data from input space to latent space (we will talk about it in a moment) is known as **Encoding** while reconstructing from latent space to input space known as **Decoding**. Following is the architecture of basic process in the AE.

<img src='1_Autoencoder_structure.png' width=500>

So as you can see an autoencoder basically a combination of two different neural networks with different responsibilities. One is responsible to encode the input data and other one reconstruct the input again.

This what the basics behind the AE. Now you may ask; why this reconstruction of input is important? What is the significance of it? To understand this, we first need to understand the basic working behind it. Let's start with it.

## 2.1.1 What is encoding and decoding ?

I know the word **Latent Space** may bother you and I appreciate your patience for keeping calm all the time. In simple English **latent means hidden**; and space is just a place or a state of the data. Guys remember this word latent space, because as the future data scientists whenever you will work with artificial neural nets, you will see this word in every literature. So what is the latent space actually is? 

Let's try to understand this whole thing with an example of matrix multiplication. Following figure shows one of the way for encoding and decoding the data.

<img src='2_Encoder_Decoder.png' width=700>

Well, If you look closely, you will understand whole process. 

We will multiply input data with a weight matrix so that it look different then actual. This is what encoding is and this data will be known as latent data. Now we can use another weight matrix and multiply it with latent data to reconstruct the original data. This process known as decoding.

Let's clarify this with an example of matrix multiplication.

Suppose we have an input matrix as:

**inp** = $\begin{bmatrix}1.5 & 2.5\\3.5 & 4.5\end{bmatrix}$

and a weight matrix as:

**weights** = $\begin{bmatrix}2 & 3\\5 & 2\end{bmatrix}$

Now when we multiply weights and input together, we will get:

**latent** = $\begin{bmatrix}2 & 3\\5 & 2\end{bmatrix}$  $\begin{bmatrix}1.5 & 2.5\\3.5 & 4.5\end{bmatrix}$

**latent** = $\begin{bmatrix}13.5 & 18.5\\14.5 & 21.5\end{bmatrix}$

So As you can see when we multiply weights and input we get a different matrix at the output which looks completely different than our input ($inp$). Voila! We have **encoded** our input successfully. Now the matrix at the output ($latent$) is what the latent space is. 

Our next task is to reconstruct $inp$ for doing it we need to find inverse of weight matrix and then we will multiply it with the $latent$ output. Let's see.

**recon** = $\begin{bmatrix}-0.1818 & 0.2727\\0.4545 & -0.1818\end{bmatrix}$  $\begin{bmatrix}13.5 & 18.5\\14.5 & 21.5\end{bmatrix}$

And the output is:

**recon** = $\begin{bmatrix}1.5 & 2.5\\3.5 & 4.5\end{bmatrix}$

Thats it! we have recovered the original input from the latent space. That means we have successfully created a **decoder**!.

The operations we have performed here are quite simple. In case of an actual AE these operations are much complex than this. We will learn about this in some moments. And by the way, you can implement above steps using following snippet

```python
import numpy as np

inp = np.array([[1.5,2.5],[3.5,4.5]])
print("\nInput")
print(inp)

weights = np.array([[2,3],[5,2]])
print("\nWeights")
print(weights)

latent = weights.dot(inp)
print("\nLatent")
print(latent)

print("\nInverse")
print(np.linalg.inv(weights))

recon = np.linalg.inv(weights).dot(latent)
print("\nReconstruction")
print(recon)
```

Everything looks perfect! Now you may ask; how the hell you will get an appropriate weight matrix to encode and decode the input data? Well that's where machine learning algorithms kicks in. We will learn these weight matrix using Back-propagation and stochastic gradient descent algorithms. This process known as training of the AE.

Here you need to understand that there will always be a certain degree of mis match between input data and its reconstruction because in practical datasets it is quite difficult to achieve perfect convergence. So we always quantify the performance of an
AE using its reconstruct error. 

To find out the reconstruct error we simply subtract input data with the reconstructed one and find out the difference between the two. Less error shows good reconstruction and vice versa. 

Now before going into more detail of it let's understand how this encoding and decoding can helps us in implementing various useful applications.

For example if input dimension of data is N X 100 (Here N is the number of samples) and weight matrix have 100 X 20 elements; then the latent space will also be reduced to the dimension of the weight matrix i.e. N X 20. Although the resultant compression will be a lossy compression it could very helpful in many applications. Following figure shows basic process of data compression using AE. # 2.2 Types of Autoencoder (AE)

So now we know what is an autoencoder and how useful it could be for different application domains. Now before moving towards implementation of AE, let's briefly discuss about what kind of autoencoders are being used for different application types.

Broadly there are 3 kind of autoencoders.

1. Vanilla autoencoder.
2. Variational autoencoder.
3. Convolutional autoencoders.

Let's discuss about each in short.

**Vanilla Autoencoder.** is a very basic form of AE. It is a simple feed forward neural net with 1 hidden layer, 1 input and 1 output layer (total 3 layers). In this architecture input layer takes inputs, hidden layer encode the input data and output layer decode it into same dimension as the input. 

following is the architecture of vanilla AE.

<img src='6_Vanilla-Autoencoder.png' width=400>

*image source: https://pythonmachinelearning.pro/all-about-autoencoders/*

We will implement the vanilla AE in some moments.

** Variational Autoencoder (VAE)** is an algorithm from generative models class. It can generate new data from the learned representation. It can be used to generate new images instead only reconstruction of learned data. We will discuss variational autoencoder for image generation in a different chapter. Basically in a VAE hidden layer (Encoder) generates mean and standard deviation of the input distribution (q(z|x), where X is input data and Z is latent variable) instead of data. While decoder takes these value and draws new samples from the distribution (p(x|z)) defined by output of encoder. This helps in generation of images using the distributions learned from the training data.

I know guys this process is bit complex. So I will discuss the whole topic of VAE and its application in a separate chapter so have patience till you read that. for now, following is the architecture of a VAE.

** Convolutional Autoencoder (CAE)** are specifically designed to work with image data. these AE consist of convolutional layers instead of fully connected layers to learn and process the features on the dataset. CAEs are generally used for image compression and denoising; We can also use a variational CAE for image generation task similar to a GAN (as discussed in previous chapter). # 2.3 Building an Autoencoder

In this chapter we will build a vanilla AE and a CAE. We will start with vanilla AE and try to understand the under lying concepts of building an AE. So without wasting our time let's start the implementation of our first AE.

## 2.3.1 Image generation using AE

We will work on image generation using our vanilla AE. So as we have discussed in previous section; Vanilla AE consists 3 layers in total; one hidden layer an two others are input and output layers. We will try to learn representation on MNIST hand written digit data for simple and clean understanding. We will use Keras with TensorFlow backend to implement our network. Let's start with it.

Following will be the steps we will implement one by one:

- Load data set.
- Create training and testing set from the data.
- Create a neural network with three layers.
- Train the neural network
- Use trained network for digit reconstruction.

Let's write the python code for above mentioned steps.

```python
# Let's start with importing the data set
from keras.datasets import mnist
import numpy as np
(x_train, _), (x_test, _) = mnist.load_data()
```

```python
# Let's quickly look at the digit image
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(x_train[25])
```

Great! we have successfully downloaded the data set. Once the data set downloaded from the amazon server; It's time to rescale the data set. We will rescale this data between 0 and 1 by dividing each pixel of the image by 255. We will also reshape the data into 1D vector of length 784 (28X28) as we are implementing a simple autoencoder.

```python
# Let's divide the data by 255
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Here we will reshape the images into 1D vector
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

# Let's see how well we have done it
print (x_train.shape)
print (x_test.shape)
```

Well all looks perfect. Our next task is to create a neural network which can learn from training data and then reconstruct the digits. Following will be our code to do this.

```python
# We will use keras to create the model
from keras.layers import Input, Dense
from keras.models import Model

# Here we define size of the latent features
encoding_dim = 64  

# This is our AE's Input layer
input_img = Input(shape=(784,))

# "encoded" is the encoded representation of the input
# This will be the Hidden layer
encoded = Dense(encoding_dim, activation='relu')(input_img)

# "decoded" is the lossy reconstruction of the input
# This will be the Output layer of the AE
# We have used sigmoid as the activation function for the last layer. 
# It will generate output in range of 0 and 1
decoded = Dense(784, activation='sigmoid')(encoded)

# Here we connect all three layers into one model
autoencoder = Model(input_img, decoded)

autoencoder.summary()
```

So we are ready to train our autoencoder model. Now before start training we will need to choose a proper loss function which will be optimize by the optimizer. Mean square error (MSE) is the best choice for autoencoders, as AE is a regression network MSE is well suited function for the task.  We will use SGD with back-propagation to optimize the loss fucntion by changing the weights. We will use ADAM for optimization.

```python
# Let's call the optimizer
from keras.optimizers import adam
from keras.callbacks import ModelCheckpoint

# Path to save the model onto local disk
path2save  = 'SimpleAE.h5'

# Define callback so we can control when to save the model.
cb = ModelCheckpoint(path2save,monitor='val_loss',verbose=2,save_best_only=True,mode='auto')

#In following lines we will compile the model
opt = adam(lr=1e-3)
autoencoder.compile(optimizer=opt,metrics=['accuracy'],loss='mse')

# It's time to train the model.
# We will use 20 % data as validation of the model.
# As this is an AE input and labels will be same.
autoencoder.fit(x_train,x_train,epochs=50,batch_size=1024,validation_split=0.2)
```

Great! validation loss of the model is decreasing in each epochs and thats what we want from our AE. Now let's quickly generate reconstructions from the trained AE and let's see how well it is performing on the test data.

```python
# Following line will feed test data into the AE and will get 
# reconstruction at the output
decoded_imgs = autoencoder.predict(x_test)

# We will use Matplotlib to visualize what is happening
import matplotlib.pyplot as plt

n = 10  # We will display 10 random digits
plt.figure(figsize=(20, 4))
id = 0
for i in range(10,20):
    # display original
    ax = plt.subplot(2, n, id + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, id + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    id+=1
plt.show()
```

Perfect! Well, our AE learned digits quite well. In above figure first row shows actual digit values and second row shows respected reconstructions. As you can see it is able to generate the reconstructions with very minimal error. Congratulations guys, you have just successfully trained your first AE.

As you can see digits are bit blurry at the reconstruction, this is due to lossy compression of the data. But don't worry we can utilize this drawback for one very interesting application i.e. Image Denoising! How? We will see next.

## 2.3.2.1 Image De-noising using vanilla AE

In previous section we had learned how to implement an AE with one hidden layer. Can we use it for image De-noising? Let's find it out quickly.

As we do not have any corrupted data with us; we will corrupt our data manually and train our network with uncorrupted data as image labels.

We will take following steps to implement the AE

1. Load data set
2. Corrupt data set
3. Train Vanilla AE for image De-noising
4. Test AE on the test images.

Let's go through each step one by one.

Following code snippet will load the data and corrupt it with the random noise.

```python
# We will use MNIST digits data set for this task
from keras.datasets import mnist
import numpy as np

# Load the training and test data set 
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize it between 0 and 1
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Reshape data set such that first dimension will be number of channels followed by height and width of images
x_train = np.reshape(x_train, (len(x_train), 1,28, 28))  
x_test = np.reshape(x_test, (len(x_test), 1,28, 28))  

# Add random gaussian noise to the images
noise_factor = 0.5
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape) 
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape) 

# Rescale data between 0 to 1 (Noise addition may change the grey scale)
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

Let's quickly look at some of the training images

```python
import matplotlib.pyplot as plt
n = 10
plt.figure(figsize=(20, 2))
for i in range(1,n):
    ax = plt.subplot(1, n, i)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

Oh My GOD! What the hell had happened to the images. Can you tell me 6th digit from above row? It's difficult. So what do you think; can we still do it?

Let's find it out.

Create our AE's model as we had created earlier

```python
# We will use keras to create the model
from keras.layers import Input, Dense
from keras.models import Model

# Here we define size of the latent features
encoding_dim = 64  

# This is our AE's Input layer
input_img = Input(shape=(784,))

# "encoded" is the encoded representation of the input
# This will be the Hidden layer
encoded = Dense(encoding_dim, activation='relu')(input_img)

# "decoded" is the lossy reconstruction of the input
# This will be the Output layer of the AE
# We have used sigmoid as the activation function for the last layer. 
# It will generate output in range of 0 and 1
decoded = Dense(784, activation='sigmoid')(encoded)

# Here we connect all three layers into one model
autoencoder = Model(input_img, decoded)

autoencoder.summary()
```

Now we are ready to train our autoencoder model. Before start training we will need to choose a proper loss function which will be optimize by the optimizer. Mean square error (MSE) is the best choice for autoencoders, as AE is a regression network MSE is well suited function for the task.  We will use SGD with back-propagation to optimize the loss fucntion by changing the weights. We will use ADAM for optimization.

```python
# Let's call the optimizer
from keras.optimizers import adam
from keras.callbacks import ModelCheckpoint

# Path to save the model onto local disk
path2save  = 'SimpleAE_Noise.h5'

# Define callback so we can control when to save the model.
cb = ModelCheckpoint(path2save,monitor='val_loss',verbose=2,save_best_only=True,mode='auto')

#In following lines we will compile the model
opt = adam(lr=1e-3)
autoencoder.compile(optimizer=opt,metrics=['accuracy'],loss='mse')

# It's time to train the model.
# We will use 20 % data as validation of the model.
# As this is an AE input and labels will be same.
# Here we will reshape the images into 1D vector
x_train_1D = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test_1D = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

# Here we will reshape the images into 1D vector
x_train_noisy_1D = x_train_noisy.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test_noisy_1D = x_test_noisy.reshape((len(x_test), np.prod(x_test.shape[1:])))


autoencoder.fit(x_train_noisy_1D,x_train_1D,epochs=50,batch_size=1024,validation_split=0.2,callbacks=[cb])
```

```python
# Following line will feed test data into the AE and will get 
# reconstruction at the output
decoded_imgs = autoencoder.predict(x_test_noisy_1D)

# We will use Matplotlib to visualize what is happening
import matplotlib.pyplot as plt

n = 10  # We will display 10 random digits
plt.figure(figsize=(20, 4))
id = 0
for i in range(10,20):
    # display original
    ax = plt.subplot(2, n, id + 1)
    plt.imshow(x_test_noisy_1D[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, id + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    id+=1
plt.show()
```

HaHa! Great! Wonderful! can you see that? We had done it. Our AE can successfully remove noise from the test data. It perform great. If you look closely it is very difficult to tell what are the digits at number 6 and 9 at our AE has clear the picture for us. But these images are very blurry. What do you think? In this section we will learn about CAE, which are specifically designed for image related analysis. There are different variants of CAE are available like vanilla CAE, Variational CAE, Sparse CAE etc. We will limit our discussion on vanilla CAE in this chapter.

For your reminder following is the architecture of a CAE.

<img src='8_CAE.png' width=650>

*image source: https://hackernoon.com/autoencoders-deep-learning-bits-1-11731e200694*

As you can see in above figure, it consist of various convolutional layers to generate latent features. These kind of architecture also known as fully convolutional neural networks (FCN) as they don't have any fully connected layers in the architecture. But why convolutional neural networks are good to use for images.

Well a convolutional neural has three main concepts which make them best suitable for computer vision applications.

1. Local receptive fields
2. Shared weights and biases.
3. Pooling Layers

Let's talk about these terms in brief.

### Local receptive fields

This helps to utilize the local information of the regions in the image, as an image is nothing but a group of regions, similarly a region in an image is nothing but a group of similar pixels. So when working with CNN our networks can learn local region based information quite well. We'll connect the input pixels to a layer of hidden neurons which is also known as filters. But here we will not connect each image pixel to each hidden neuron. Instead, we keep these connections in a small area only. 

So each neuron from the hidden layer will be connected to a small region (local field) of the input pixels, for example, let’s consider a 5×5 region, which corresponds to 25 input pixels. So, for a specific hidden neuron, we may have connections which look like this:

<img src='5_LocalReceptive.png' width=400>

The region shown in blue is the receptive field and the same receptive field is also known as **convolutional filter**. We will create this kind of receptive field around each pixel in the image and perform a sum of product using specific coefficients (Weights). These coefficients will be learned during the training process and it will be responsible for extraction of specific features from different location on the image. We will learn about these weights in next paragraph.

### Shared Weights and Biases

The concept of shared weights and biases lies somewhere in the operation we have discussed above. As we have seen a single filter spans over the entire image to generate the output. It is known as the weight and bias sharing. Here unlike the plain feed forward network every filter matrix shares same pixel information. 

What it means actually? Well suppose a filter of the first layer can be used for horizontal edge detection at one place. Then you can use the same filter to calculate the similar edge features at different locations on the image. This information sharing helps in reusing of the filters which leads to a computation efficient solution. This information sharing ability also helps convolutional neural networks to achieve translations invariance. So if you have an image of a face which could be anywhere in the image and your classifier can still classify it as the face image. 

Similarly our network tries to detect other kind of features at different location on the image; this means each receptive field will have its own output which we called the feature map. Now to get the solution of an image classification problem you need to have lot of convolutional filters at each hidden layer which can detect more abstract features at each level (layer) in the network. Let's have a sneak pick at some filters which we will learn later in the chapter.

<img src='6_EdgeFilters.jpg' width=400>

Figure above shows the filters from the first layer trained on LeNet architecture. As you can see clearly these 25 features are looks like edge detectors which can detect different kind (orientation) of edges on the input image.
So the whole operation performed above is also known as Convolution and thus these neural networks got the name convolutional neural networks and these filters are known as Convolution filters.

We can summarize the operation of convolution using a 5X5 filter in following equation.

\begin{align}
\sigma\left(b+ \sum_{l=0}^4 \sum_{m=0}^4 w_{l,m} a_{j+l,k+m} \right) \\
\end{align}

So here $w_{l,m}$ is an element of the weight matrix (remember these are the 2D filters), $a_{j+l,k+m}$ is the input pixel, $b$ is the bias and $\sigma$ is the activation function applied on the output.

Now let's take a look on pooling operation.

###  Pooling Layers 

As receptive fields and shared weights idea pooling is also an important aspect of the CNN architecture. It helps to reduce the noisy activations for the descendant layers by selecting different kind of pooling operation. Pooling layer compress the dense predictions and reduce the size of the input. This operation is unlike the strided convolutions. In pooling there is no learnable parameter available so it does not have any weight matrix. It selects activations from the previous layers from specific indices and generates the compressed output. Following figure will give you better understanding of the pooling operation.

<img src='7_pooling.png' width=400>

Figure above shows a 2X2 max pooling operation on a learned feature map. As you can see pooling layer creates a 2X2 pixels region over the input feature map and selects maximum value of the region and put it on the output map. You can choose minimum or average of the region as the output value. So when you apply max pooling on a 14X14 region the output will be of 7X7 pixels. By selecting the maximum or average value of the region it picks only the strongest activations which are most likely to be part of region boundaries.

Pooling operation also helps to learn more abstract features. This helps our network to define more refine object properties as down sampling reduces finer details from the images. This helps the network to learn more coarser features.

Now as you can see with the help of local receptive fields, shared weights and pooling a CNN can learn more complex features from images. These features could be very useful to identify different objects present in the image. We will use these features to create complex representation to generate descriptions for images.

I think we had learned enough to go ahead and build a CNN which can actually reconstructs digits from noisy data. One more important concept we will be using is Up-sampling of latent features. As we are working with Encoder-Decoder architecture, we need to retain the image shape to get proper reconstructions. To do this we need to up-sample the latent features so it can be matches with image's actual dimensions at the end. We will a bilinear interpolation to perform this task.

Our architecture consist of following architecture.

- 5 Convolutional layers with rectified linear units (ReLU) activation.
- 2 Pooling layers; we will use Max-pooling.
- Final convolution layer with sigmoid neuron to generate output probabilities between 0 and 1.

Let's create our model in following code snippet.

```python
from keras.layers import Input,Conv2D,MaxPooling2D,UpSampling2D
from keras.models import Model
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam

def DenoisingModel(model_weights=None):
    
    input_img = Input(shape=(1,28, 28))  # adapt this if using `channels_first` image data format

    x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)    
    x = MaxPooling2D((2, 2), padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    encoded = MaxPooling2D((2, 2), padding='same')(x)

    # at this point the representation is (7, 7, 32)

    x = Conv2D(32, (3, 3), activation='relu', padding='same')(encoded)
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = UpSampling2D((2, 2))(x)
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

    autoencoder = Model(input_img, decoded)

    if model_weights is not None:
        autoencoder.load_weights(model_weights)
        
    return autoencoder    
```

Now we need to compile the model and train it with noisy data as input and clean data as labeled. We will do this in following code snippet.

```python
autoencoder = DenoisingModel()
autoencoder.summary()
```

Let's set some model checkpoint to store our model on disk. We will monitor the validation loss of the model to evaluate its performance.

```python
path2save = './DenoisingAE.h5'
cb = ModelCheckpoint(path2save,monitor='val_loss',verbose=0, save_best_only=True,mode='auto')

optimize = Adam(lr=1e-2)
autoencoder.compile(optimizer=optimize,loss='categorical_crossentropy',metrics=['accuracy'],sample_weight_mode='auto')
autoencoder.fit(x_train_noisy,x_train,batch_size=1024,epochs=50,validation_split=0.2,callbacks=[cb])
```

```python
autoencoder = DenoisingModel(path2save)

# Following line will feed test data into the AE and will get 
# reconstruction at the output
decoded_imgs = autoencoder.predict(x_test_noisy)

# We will use Matplotlib to visualize what is happening
import matplotlib.pyplot as plt

n = 10  # We will display 10 random digits
plt.figure(figsize=(20, 4))
id = 0
for i in range(10,20):
    # display original
    ax = plt.subplot(2, n, id + 1)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, id + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    id+=1
plt.show()
```
