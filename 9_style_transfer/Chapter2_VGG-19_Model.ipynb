## Unit 1. Evolution of CNN models 

To just give a brief idea on **Convolutional Neural Network** (**CNN** or **ConvNet**), they are advanced versions of multi-layer neural networks, which can recognise patterns from images. The backbone of these networks are convolutional filters with depth and series of activations. Let's not dig deep in to the CNNs since they may have already been covered in the other chapters previously.

The ImageNet project is a large visual database of labelled images used for visual object recognition research. It runs an annual software contest **ImageNet Large Scale Visual Recognition Challenge**(**ILSVRC**), where researchers, industries compete to correctly classify and detect objects and scenes from the Imagenet dataset. The evolution of CNN models are hugely attributed to the state of the art papers (models) which achieved highest accuracy in ILSVRC each year. 

![](https://bitbucket.org/ga_learning/style_transfer/raw/48db127941e04f2f66cdfd6a3ccaca9ed888d598/markdown_images/Evolution_CNN1.png)


It all started with LeNet (1998). In last 15 years, many deep models performed well in the ILSVRC. Below is a list of the famous models along with a very brief description 


### LeNet in 1998

LeNet is a 7-level convolutional network by LeCun that classifies digits and used by several banks to recognise hand-written numbers on cheques digitized in 32x32 pixel greyscale images. The ability to process higher resolution images requires larger and more convolutional layers, so this technique is constrained by the availability of computing resources.


![](https://bitbucket.org/ga_learning/style_transfer/raw/48db127941e04f2f66cdfd6a3ccaca9ed888d598/markdown_images/lenet.png)
 


### AlexNet in 2012

In the course of years there was a surge in computational power which enabled the researchers to go ahead of the LeNet. This led to the birth of deep learning and gave rise to ground breaking models. 

AlexNet is considered to be the first model which demonstrated the power of Deep-CNNs when it won the ImageNet challenge in 2012. AlexNet is trained on ImageNet and outperformed all the entries that year with a top error rate of 15.3% . It was a major improvement with the next best entry getting only 26.2% top 5 test error rate. Compared to modern architectures, a relatively simple layout was used.

### ZFNet in 2013

ZFNet is a modified version of AlexNet which gives a better accuracy.

One major difference in the approaches was that ZF Net used 7x7 sized filters whereas AlexNet used 11x11 filters. The intuition behind this is that by using bigger filters we were losing a lot of pixel information, which we can retain by having smaller filter sizes in the earlier conv layers. The number of filters increase as we go deeper. This network also used ReLUs for their activation and trained using batch stochastic gradient descent.

### VGG-Net in 2014

The idea of VGG was submitted in 2013 and it was a runner up in the ImageNet contest in 2014. It was created by the **Visual Geometry Group, Oxford University**, hence the name VGG-Net.

VGG Net used 3x3 filters compared to 11x11 filters in AlexNet and 7x7 in ZFNet. The authors give the intuition behind this that having two 3x3 filters gives an effective receptive field of 5x5, and three 3x3 filters give a receptive field of 7x7 filters, but using this we can reduce the number of parameters in the network. VGG-Net is famous for it's simple architecture. 

![](https://bitbucket.org/ga_learning/style_transfer/raw/48db127941e04f2f66cdfd6a3ccaca9ed888d598/markdown_images/vgg-16_19.png)



### GoogleNet in 2014 (InceptionNet)

In 2014, several great models were developed like VGG but the winner of the ImageNet contest was GoogleNet.

GoogLeNet proposed a module called the inception modules which includes skip connections in the network forming a mini module and this module is repeated throughout the network.

GoogLeNet uses 9 inception module and it eliminates all fully connected layers using average pooling to go from 7x7x1024 to 1x1x1024. This saves a lot of parameters.

![](https://bitbucket.org/ga_learning/style_transfer/raw/48db127941e04f2f66cdfd6a3ccaca9ed888d598/markdown_images/googlenet-1.png)



### ResNet in 2015

This is the modern day highly efficient deep-network designed by Microsoft Research. There are 152 layers in the Microsoft ResNet. The authors showed empirically that if you keep on adding layers, the error rate should keep on decreasing in contrast to “plain nets” where adding a few layers resulted in higher errors both while training and testing.

The reason why residual blocks improve classification is due to the presence of skip connections between the layers. These skip connections add the outputs from previous layers to the outputs of next layers. This helps to train much deeper networks than what was previously possible and hence results in better accuracy. 

![](https://bitbucket.org/ga_learning/style_transfer/raw/48db127941e04f2f66cdfd6a3ccaca9ed888d598/markdown_images/resnet.png)

## Unit 2. VGG-19 Architecture 

VGG-Net is one of the most preferred choice in the deeplearning community for extracting features from images. 

The main reasons are : 
* It has a very straight-forward and uniform arrangement of the layers making it very simple to tap the intermediate outputs necessary for style transfer 
* Better models like ResNet, have complex architecture units like residual blocks and skip connections make it complicated to tap the intermediate outputs
* The weights configuration of the VGGNet is publicly available. It is used as baseline feature extractor in many applications 

The VGG network architecture was introduced by Simonyan and Zisserman in their 2014 paper, **Very Deep Convolutional Networks for Large Scale Image Recognition** [https://arxiv.org/abs/1409.1556]. The original VGGNet consists of 16 convolutional layers (**VGG-16**) and the more deeper version has 19 convolutional layers (**VGG-19**). It achieved the top-5 accuracy of 92.3 % on ImageNet.

Inspired from Alex-net, this network is known for it's simplicity. As opposed to large kernels like  **11x11** and **7x7** used before, VGG-net's main highlight is using smaller (**3×3**) convolutional layers stacked on top of each other in increasing depth. Maxpooling is used to handle the increasing volume size. It starts with a depth of 64 and increases by factor of 2 after each pooling till it reaches 512 . Then, three fully-connected layers, two with 4,096 nodes and one with 1000 nodes are used followed by a softmax classifier. ReLU is used as activation.

The intuition is that more convolution layers with smaller kernels stacked together increases the effective **receptive field** of the network without using high number of parmeters. The receptive field is the effective area of input image on which output depends. Also with more filters the non-linearity increase and so does the depth which helps us learn more complex features better.

This idea is inspired from the fact that, the number of parameters grows quadratically with kernel size as per the equation, 

\begin{align}
\ parameters & = kernel^2 * Cin * Cout \\
\end{align}

where,  Cin, Cout are the input and out channels of the layer.


For example : 

3 convolutional filters of 3x3 kernels stacked together give us an effective receptive field of a single convolutional filter of 7x7 kernel . Let us now calculate the parameters involved. 

For each 3x3 filter, params = 3<sup>2</sup> x Cin x Cout = **9 x Cin x Cout** 

For 3 such 3x3 filters, params = 3 x 9 x Cin x Cout  = **27 x Cin x Cout**

For 1, 7x7 filter, params = 7<sup>2</sup> x Cin x Cout = **49 x Cin x Cout** 

So using 3, 3x3 filters we have reduced the parameters by **45%** approximately compared to single 7x7 filters


This makes bigger convolution kernels not cost efficient enough. So replacing them with multiple smaller kernels would give us the same effective receptive field with much lesser parameters. 

With a given receptive field(the effective area size of input image on which output depends), multiple stacked smaller size kernel is better than the one with a larger size kernel because multiple non-linear layers increases the depth of the network which enables it to learn more complex features, and that too at a lower cost.

## Unit 3. VGG-19 in Keras

There are 2 ways to create a model for our applications

1. Build our own model from scratch 

2. Transfer Learning 

Approach 1 though a good one, needs good amount of time and efforts.It starts right from data creation and then building the architecture, training,  monitoring, testing. This will be suitable for applications which need pinpoint accuracy and deal with very unique applications. 

Approach 2 ie., transfer learning is quite popular in most of the deeplearning applications. Using transfer learning approach we can utilise the famous pretrained models for our own applications. The general practice is to strip the layers in the end (any number of layers based on applications), freeze the remaining part of the network and add our own layers to it. Training this setup on our own data, often gives amazing results in very less time since it utilizes the knowledge already learnt by the pretrained network. The main reason transfer learning is successful is due to the enormous datasets on which state of the art models are trained. The datasets like Imagenet, COCO cover huge number of real world classes and hence very powerful. 

So the advantages are clear, we will use approach 2 since it is faster and known to be reliable. Also style transfer needs us to extract features from intermediate layers and doesn't necessarily needs training a new model 

Almost all the deeplearning frameworks support transfer learning by providing pretrained models as APIs. We will now see how simple it is using Keras.

### Loading the VGG-19 model 

Keras **Applications** module helps us load famous models alongwith pre-trained weights. These models can be used for prediction, feature extraction, and fine-tuning.In Keras the weights will be downloaded automatically when instantiating a model. This is a one time process since, the downloaded weights will be stored at **~/.keras/models/** and will be loaded from that path automatically when we use the model again next time.

Detailed documentation can be found at : https://keras.io/applications/

```python
#from tensorflow.contrib.keras.api.keras.applications import vgg19
from keras.applications import vgg19 
def load_VGG19(include_top=False):
    """
     include_top: whether to include the 3 fully-connected layers at the top of the network.
     weights: one of None (random initialization) or 'imagenet' (pre-training on ImageNet)
     input_tensor: optional Keras tensor (i.e. output of layers.Input()) to use as image input for the model
     
     More options at : 
     https://keras.io/applications/#extract-features-from-an-arbitrary-intermediate-layer-with-vgg19
    """
    model = vgg19.VGG19( include_top=include_top,weights='imagenet') # input_tensor=input_tensor 
    return model 

include_top = True
vgg19_model = load_VGG19(include_top)
```

```python
vgg19_model.summary()
```

### Data preprocessing 

Since we are using VGG-19 trained on the ImageNet dataset , we need to follow the same preprocessing steps as used for ImageNet dataset for the best results.

The basic preprocessing steps employed while training on ImageNet dataset are 

* Crop the image to dimensions 224 x 224 which is standard input image size for VGG-net. Random crop is used generally during training for robustness. Centre crop would be better only if the object lies in the centre which is generally not the case. For the style transfer pipeline we won't be cropping the image but rather resizing it directly to 224 x 224 since we cannot afford to lose any part of the original image due to cropping.   
* Horizontal flip 
* Normalise the image by 255.0 to get values from 0 to 1
* Mean subtract each channel by ImageNet specfic mean values 
        Red_MEAN = 0.485 (123.68 if not normalised)
        Green_MEAN = 0.456 (116.779 if not normalised )
        Blue_MEAN = 0.406 (103.939 if not normalised )
* Normalise each channel by ImageNet specfic std deviations (Optional)
        Red_std = 0.229
        Green_std = 0.224
        Blue_std = 0.225     
        
The keras vgg19 module has it's own **preprocess_input** method which can help us do this
Please refer the function **preprocess(img)** below 

Ref : 
https://github.com/keras-team/keras-applications/blob/master/keras_applications/vgg19.py for the source code of the preprocess_input() function which also highlights the above mean and std values.

```python
import numpy as np
from tensorflow.contrib.keras.api.keras.preprocessing.image import load_img, img_to_array
from keras.applications.imagenet_utils import decode_predictions
import matplotlib.pyplot as plt
%matplotlib inline

def preprocess(img):
   assert(img.ndim == 4 ),"The input images should be of the format : batch_size x 224 x 224 x 3"
   pp_img = vgg19.preprocess_input(img.copy())
   return pp_img

# Load image and resize to 224x224x3
img_path = 'images/content_imgs/bridge.jpg'
img_orig = load_img(img_path, target_size=(224, 224))

# Since load_img() loads image in PIL format, conevert it to numpy array
img = img_to_array(img_orig)

# Restructure image dimensions to  Batch_size x  imshape format (1 x 224 x 224 x 3)
img = np.expand_dims(img, axis=0)
print("Input Image")
plt.imshow(img[0].astype('uint8'))
plt.show()

# Preprocess image
pp_img = preprocess(img)
print("Preprocessed Image")
plt.imshow(pp_img[0].astype('uint8'))
plt.show()
```

### Data deprocessing 

We will need a deprocessing function which restores the pixel values in the output image to their original values before normalization. We use the above mentioned mean values for getting the original images back.
Please refer the function **deprocess(img)** below 

We will be using both the process and deprocess functions in the upcoming chapters as well. 

```python
def deprocess(img):
    assert(img.ndim == 4 ),"The input images should be of the format : batch_size x 224 x 224 x 3"
    dp_img = np.zeros_like(img)
    batch_size = img.shape[0]
    imagenet_mean = [103.939,116.779,123.68]
    for b in range(batch_size):
        for i in range(3): #BGR
            dp_img[b,:,:,i] = img[b,:,:,i] + imagenet_mean[i]
    dp_img = dp_img[:,:,:,::-1]#Make it RGB
    return dp_img

print("Input Image")
plt.imshow(img[0].astype('uint8'))
plt.show()

print("Preprocessed Image")
plt.imshow(pp_img[0].astype('uint8'))
plt.show()

dp_img = deprocess(pp_img) #Observe the input here . It is preprocessed image
print("Deprocessed Image")
plt.imshow(dp_img[0].astype('uint8'))
plt.show()
```

###  Getting predictions from the loaded network

###  Extracting features from the loaded network 

We will explore 2 ways to access features from intermediate layers. For that we will need to know the model architecture and naming convention for each layers. This can be obtained using model.summary()

```python
vgg19_model.summary()
```

### Using Keras Backend Functionality **'K'** 

```python
from keras import backend as K
get_layer_output = K.function([vgg19_model.layers[0].input],
                                  [vgg19_model.layers[3].output])
method1_features = get_layer_output([pp_img])[0]
print(method1_features.shape)
```

### Using Keras.models.Model

```python
from keras.models import Model
layer_name = 'block1_pool'
intermediate_layer_model = Model(inputs=vgg19_model.input,
                                 outputs=vgg19_model.get_layer(layer_name).output)
method2_features = intermediate_layer_model.predict(pp_img)
print(method2_features.shape)
```

We can see that both the above methods produce the same output since the difference between the feature maps extracted is 0. 

```python
print(np.max(method2_features - method1_features))
```

### Visualising some feature maps from different stages of network

We extract feature maps at :

block1_pool  - Pool 1 (Near Begining)

block2_pool  - Pool 2

block3_pool  - Pool 3

block4_pool  - Pool 4 (Near End)

We can clearly see that the features extracted quite prominent in the start (Pool 1) and slowly become more and more specific as we move towards end of the network (Pool 4)

At Pool 1, we can easily identify the outline of the Husk, more like an edge image. Hence the initial layers are good at identifying the overall texture and objects in the image. 

At Pool 4, we can only see fragmented features which probably are finer level features extracted across 512 feature maps. The layers towards end are good at identifying the fine and specific details like eyes, ears etc.
